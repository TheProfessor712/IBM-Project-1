### Below is code for a Riemann sum over the interval [0,1]x[0,1] with an m,n partition to approximate the integral of the function $f(x,y) = 16-x^2-y^2$.

```python
V=0
m=3 #partitions along the x-axis
n=3 #partitions along the y-axis

def f(x,y):
    return 16-(x**2)-(y**2)  #Defined our function - feel free to pick another!

for i in range(0,m):
    for j in range (0,n):
        V = V + (1/m)*(1/n)*f(i/m, j/n) #adding the volumes of the approximating rectangular prisms together
        
print(V)
```
